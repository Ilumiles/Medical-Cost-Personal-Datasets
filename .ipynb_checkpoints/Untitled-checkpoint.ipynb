This is a classic **insurance charges prediction** dataset, where:

* **Target variable:** `charges` (continuous) → regression problem.
* **Features:** `age`, `sex`, `bmi`, `children`, `smoker`, `region`.

A good EDA should answer three questions:

1. What does each variable look like?
2. How do the variables relate to each other?
3. Which variables affect insurance charges the most?

# 10. Feature Importance (after encoding)

Once you've encoded the categorical variables, you can estimate which features are most informative.

For example, using mutual information:

```python
from sklearn.feature_selection import mutual_info_regression

X = encoded_df.drop("charges", axis=1)
y = encoded_df["charges"]

mi = mutual_info_regression(X, y)

pd.Series(mi, index=X.columns).sort_values(ascending=False)
```

Or after fitting a Random Forest:

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(random_state=42)
model.fit(X, y)

importance = pd.Series(model.feature_importances_, index=X.columns)
importance.sort_values(ascending=False)
```

## Expected Insights

By the end of your EDA, you should be able to answer questions like:

* Which feature has the strongest relationship with `charges`? (Typically `smoker`.)
* Does `age` increase insurance charges?
* Is higher `bmi` associated with higher charges?
* Are insurance charges right-skewed?
* Are there significant outliers in `charges`?
* Do `sex` or `region` meaningfully affect charges?
* Are there missing values or duplicate records that need cleaning?

Since you're learning machine learning, this dataset is also a good opportunity to explore **interaction effects**, such as whether BMI has a stronger impact on charges for smokers than for non-smokers:

```python
sns.scatterplot(
    data=df,
    x="bmi",
    y="charges",
    hue="smoker"
)
```

Plots like this can reveal patterns that a simple correlation matrix might miss and help you think about feature engineering before building your regression model.
