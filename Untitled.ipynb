## Univariate Analysis

```python
df.head()
```

```python
df[["age","bmi"]].describe()
```

```python
df.hist(figsize=(10,8))
plt.show()
```

```python
bins = range(15, 70, 5)

age_groups = pd.cut(df["age"], bins=bins)

age_groups.value_counts().sort_index().plot(kind="bar")

plt.xlabel("Age Group")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
columns = ["charges", "age", "bmi"]

for col in columns:
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```
